```python
import cv2
import matplotlib.pyplot as plt
import os 

img = cv2.imread('/content/Images/Angry Bird/Cax_train1.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
data_dir = 'Images'
folder_list = os.listdir(data_dir)
for folder in folder_list:
  path_dir = os.path.join(data_dir,folder)
  for img in os.listdir(path_dir):
    img_array = cv2.imread(os.path.join(path_dir,img))
    plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
    plt.show()
    break
```

```python
img_size = 30
img_array = cv2.resize(img_array,(img_size,img_size))
plt.imshow(img_array)
plt.show()
```

```python
training_data = []
def create_training_data():
  for folder in folder_list:
    path_dir = os.path.join(data_dir,folder)
    class_name = folder
    for img in os.listdir(path_dir):
      img_array = cv2.imread(os.path.join(path_dir,img)) 
      img_array = cv2.resize(img_array,(img_size,img_size))
      training_data.append([img_array,class_name])
```

```python
create_training_data()
```

```python
training_data[0]
```

```python
import numpy as np
x = []
y = []

for feature, label in training_data:
  x.append(feature)
  y.append(label)

x = np.array(x).reshape(-1, img_size, img_size, 3)
x = x/255
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder_y = encoder.fit_transform(y)
encoder_y

```

```python
from sklearn.model_selection import train_test_split

xtr, xv, ytr, yv = train_test_split(x,encoder_y, test_size = 0.1, random_state = 0)
```

```python
print(xtr.shape)
```

```python
x.shape[1:]
```

```python
from tensorflow.keras import Sequential, layers
from tensorflow.keras.layers import Flatten
```

```python

ai = Sequential()
ai.add(Flatten())
ai.add(layers.Dense(1, activation = 'sigmoid', input_shape = x.shape[1:] ))

ai.compile(optimizer = 'sgd', loss='binary_crossentropy', metrics=['accuracy'])

ai.fit(xtr, ytr, epochs = 10, validation_data = (xv,yv))
```
